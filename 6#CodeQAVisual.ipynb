```python
import pandas as pd
import random
import os  
```

```python
reivews_df = pd.read_csv("./data/reivews_df.csv")
tests_result_df = pd.read_csv("./data/tests_result_df.csv")
tests_df = pd.read_csv("./data/tests_df.csv")
filepath = "./output/"
files = os.listdir(filepath)  
```

```python
codeqa_seed = random.choice(range(len(files)//2))
answer_file = files[codeqa_seed*2]
print(" ".join(answer_file[:-4].split("+")))
answer = pd.read_csv(filepath+answer_file)
answer['fingerprint'] = answer['status_str'] + answer['errorStack_str']
# answer.head()
print()
fingerprint = answer['fingerprint'].drop_duplicates().values
for fp in fingerprint:
    print(fp)
    content_id = answer[answer.fingerprint == fp]['content'].values
    test_id = answer[answer.fingerprint == fp]['testId'].values
    for index, c in enumerate(content_id):
        print(test_id[index], content_id[index])
    print()
```
