# Lesson QA

```python
lesson_question_df = pd.read_csv("./data/lesson_question_df.csv")
# lesson_question_df
```

```python
lesson_question_df['isQues'] = lesson_question_df.apply(lambda line: line[1] == line[2], axis=1)
lesson_question_df
```

```python
def get_lessonQA_values(lessonQA_key):
    conditions = (lesson_question_df.course == lessonQA_key[0]) & (lesson_question_df.lesson == lessonQA_key[1]) \
        & (lesson_question_df.owner == lessonQA_key[2])
    return lesson_question_df[conditions].sort_values(by=['createdTime'], ascending=True)
```

```python
def print_lesson_QA(lesson_question_df):
    str_QA = ''
    
    for line in lesson_question_df.values:
#         print(int(line[-3]//1000))
        str_QA += datetime.datetime.fromtimestamp(
            int(line[-3]//1000)
        ).strftime('%Y-%m-%d %H:%M:%S')
        
        if line[-1]: str_QA += ' Q：'
        else: str_QA += ' A：'
            
        str_QA += line[1]
        str_QA += ' '
        str_QA += line[-2]
        str_QA += '\n'
        
    return str_QA
```

## S2 根据CodeQA key获取所有的test_id

```python
def get_testids_values(codeQA_key):
    conditions = (tests_df.course == codeQA_key[0]) & (tests_df.lesson == codeQA_key[1]) \
        & (tests_df.question == codeQA_key[2]) & (tests_df.owner == codeQA_key[3])
    return tests_df[conditions].sort_values(by=['createdTime'], ascending=True)['id'].drop_duplicates().values
```
