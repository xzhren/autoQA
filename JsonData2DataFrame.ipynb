```python
import numpy as np
import pandas as pd
import json
import time
from pprint import pprint
```

```python
def get_course_lesson_map(json_data):
    course_lesson = {}
    for key in json_data:
        course = json_data[key]['course']
        lesson = json_data[key]['lesson']
        if course in course_lesson.keys():
            if lesson not in course_lesson[course]:
                course_lesson[course].append(lesson)
        else:
            course_lesson[course] = [lesson]
    return course_lesson
```

```python
def to_dataframe_lesson_questions(json_data):
    all_list = []
    for key in json_data:
        item = json_data[key]
        line = []
        line.append(item['id'])
        line.append(item['creator'])
        line.append(item['owner'])
        line.append(item['course'])
        line.append(item['lesson'])
        line.append(item['createdTime'])
        line.append(item['content'])
        all_list.append(line)
    return pd.DataFrame(all_list, columns=['id', 'creator', 'owner', 'course', 'lesson', 'createdTime', 'content'])
```

```python
with open('./data/lesson-questions.json', encoding="utf-8") as data_file:    
    data = json.load(data_file)
    
for key in list(data.keys())[:3]:
    print(key)
#     print(data[key])
    pprint(data[key])
    time.sleep(1)
    
# get_course_lesson_map(data)
lesson_question_df = to_dataframe_lesson_questions(data)
lesson_question_df
```

```python
lesson_question_df.to_csv("./data/lesson_question_df.csv", index=False, encoding="utf-8")
pd.read_csv("./data/lesson_question_df.csv")
```

```python
def to_dataframe_reivews(json_data):
    all_list = []
    for key in json_data:
        item = json_data[key]
        line = []
        line.append(item['id'])
        line.append(item['commitId'])
        line.append(item['creator'])
        line.append(item['owner'])
        line.append(item['course'])
        line.append(item['lesson'])
        line.append(item['lineNum'])
        line.append(item['path'])
        line.append(item['createdTime'])
        line.append(item['content'])
        all_list.append(line)
    return pd.DataFrame(all_list, columns=['id', 'commitId', 'creator', 'owner', 'course', 'lesson', 'lineNum', 'path', 'createdTime', 'content'])
```

```python
with open('./data/reivews.json', encoding="utf-8") as data_file:    
    data = json.load(data_file)
    
# for key in list(data.keys())[:3]:
# #     print(data[key])
#     pprint(data[key])
#     time.sleep(1)
    
# get_course_lesson_map(data)
reivews_df = to_dataframe_reivews(data)
reivews_df
```

```python
reivews_df.to_csv("./data/reivews_df.csv", index=False, encoding="utf-8")
pd.read_csv("./data/reivews_df.csv")
```

```python
def to_dataframe_tests_result(json_data):
    all_list = []
    for key in json_data:
        item = json_data[key]
        line = []
        line.append(item['id'])
        line.append(item['testId'])
        line.append(item['title'])
        line.append(item['result'])
        line.append(item['duration'])
        line.append(item['errMessage'])
        line.append(item['errStack'])
        all_list.append(line)
    return pd.DataFrame(all_list, columns=['id', 'testId', 'title', 'result', 'duration', 'errMessage', 'errStack'])
```

```python
with open('./data/test-results.json', encoding="utf-8") as data_file:    
    data = json.load(data_file)
    
for key in list(data.keys())[:3]:
#     print(data[key])
    pprint(data[key])
    time.sleep(1)

# get_course_lesson_map(data)
tests_result_df = to_dataframe_tests_result(data)
tests_result_df
```

```python
tests_result_df.to_csv("./data/tests_result_df.csv", index=False, encoding="utf-8")
pd.read_csv("./data/tests_result_df.csv")
```

```python
def to_dataframe_tests(json_data):
    all_list = []
    for key in json_data:
        item = json_data[key]
        line = []
        line.append(item['id'])
        line.append(item['tester'])
        line.append(item['course'])
        line.append(item['lesson'])
        line.append(item['question'])
        line.append(item['status'])
        line.append(item['createdTime'])
        line.append(item['compileErrStack'])
        all_list.append(line)
    return pd.DataFrame(all_list, columns=['id', 'tester', 'course', 'lesson', 'question', 'status', 'createdTime', 'compileErrStack'])
```

```python
with open('./data/tests.json', encoding="utf-8") as data_file:    
    data = json.load(data_file)
    
for key in list(data.keys())[:3]:
#     print(data[key])
    pprint(data[key])
    time.sleep(1)
    
# get_course_lesson_map(data)
tests_df = to_dataframe_tests(data)
tests_df
```

```python
tests_df.to_csv("./data/tests_df.csv", index=False, encoding="utf-8")
pd.read_csv("./data/tests_df.csv")
```

```python
merge_tests_df = tests_df.merge(tests_result_df, left_on='id', right_on='testId', how='outer')
merge_tests_df
```

```python
merge_tests_df.to_csv("./data/merge_tests_df.csv", index=False, encoding="utf-8")
pd.read_csv("./data/merge_tests_df.csv")
```
